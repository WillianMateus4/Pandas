### Início

```python
import matplotlib as plt
import pandas as pd

# df = dataframe = tabela
# sep = é o tipo de separador que é utilizado na planilha, que neste exemplo é o ponto e vírgula (;)
vendas_df = pd.read_csv(r'..\dados\chocolate\Vendas_Chocolate.csv', sep=';')

# Visualizar as informações em formato de tabela
display(vendas_df)
```

```python
# Visualizar todas as informações do dataframe
print(vendas_df.info())
```

```python
# Cria uma descrição estatística das colunas númericas
display(vendas_df.describe())

# Quartil é a divisão de um conjunto em quatro partes sendo que:
# 25% é o 1º Quartil contendo os menores valores
# 50% é o 2º Quartil contendo os menores valores e assim sucessivamente
```

### Tratamento de Dados

```python
# A coluna 'Data de Hoje' possui muitos valores vazios e não agregará para a análise das vendas
vendas_df = vendas_df.drop(columns='Data de Hoje')
```

```python
# Exclui as linhas com valores vazios
vendas_df = vendas_df.dropna()
```

```python
# Alterando o formato de data passando como parâmetro 'dayfirst=True', onde considera o dia como início da data, como está no nosso dataframe
vendas_df['Data'] = pd.to_datetime(vendas_df['Data'], dayfirst=True)
```

```python
# Substitui o '$' e a ',' por espaço vazio
vendas_df['Valor'] = vendas_df['Valor'].str.replace('$', '').str.replace(',', '')

# Altera o formato de 'str' para 'float'
vendas_df['Valor'] = vendas_df['Valor'].astype(float)
```

```python
# Remove valores duplicados
vendas_df = vendas_df.drop_duplicates()

# Remove valores duplicados em colunas específicas
# vendas_df = vendas_df.drop_duplicates(subset=['Produto', 'Valor', 'Data'])
```

```python
# Reseta o index e 'drop=True' faz com que não crie uma coluna com o antigo index
vendas_df = vendas_df.reset_index(drop=True)
```

```python
print(vendas_df.info())
```

```python
display(vendas_df)
```

### Análises

##### Análise de crescimento de vendas

```python
import plotly.express as px
```

```python
# vendas_dia recebe um dataframe com apenas as colunas 'Data' e 'Valor'
vendas_dia = vendas_df[['Data', 'Valor']]
```

```python
# Agrupar as vendas por data e 'as_index=False' para a data não se tornar index e poder executar o gráfico corretamente
vendas_dia = vendas_dia.groupby('Data', as_index=False).sum()
```

```python
grafico_data = px.line(vendas_dia, x='Data', y='Valor')
grafico_data.show()
```

```python
display(vendas_dia)
```

```python
# Feature Engineering

# Adiciona uma nova coluna chamada 'Mes' e com o formato de Mês - Ano (Ex: 2022-01)
vendas_df['Mes'] = vendas_df['Data'].dt.to_period('M')

# Adiciona uma nova coluna chamada 'Ano'
vendas_df['Ano'] = vendas_df['Data'].dt.year
```

```python
venda_mes = vendas_df[['Mes', 'Valor']]
venda_mes = venda_mes.groupby('Mes', as_index=False).sum()
```

```python
# No 'x' é preciso passar a data por inteiro ou valor númerico, neste caso é necessário transformar em 'str'
grafico_mes = px.line(venda_mes, x=venda_mes['Mes'].astype(str), y='Valor')
grafico_mes.show()
```

```python
# Gráfico de barra vertical
grafico_mes = px.bar(venda_mes, x=venda_mes['Mes'].astype(str), y='Valor')
grafico_mes.show()
```

#### Análise de volume e ticket médio

```python
# vendas_anuais recebe um dataframe com apenas as colunas 'Ano' e 'Valor'
vendas_anuais = vendas_df[['Ano', 'Valor']]

# Agrupa os valores por ano
vendas_anuais = vendas_anuais.groupby('Ano').sum()
```

```python
display(vendas_anuais)
```

```python
# Formatando a visualização do valor
display(vendas_anuais.style.format({
    'Valor': '$ {:,.2f}' # $1,000,000.00
}))
```

```python
vendas_anuais = vendas_df[['Ano', 'Valor', 'Caixas Enviadas']]

vendas_anuais = vendas_anuais.groupby('Ano', as_index=False).sum()

vendas_anuais['Valor Medio por Caixa'] = vendas_anuais['Valor'] / vendas_anuais['Caixas Enviadas']
```

```python
display(vendas_anuais.style.format({
    'Valor': '${:,.2f}', # $1,000,000.00
    'Caixas Enviadas': '{:,.0f}', # 1,000 (torna a visualização de um valor inteiro)
    'Valor Medio por Caixa': '${:,.2f}'
}))
```

```python
vendas_anuais['Valor YoY'] = vendas_anuais['Valor'].pct_change()
vendas_anuais['Caixas Enviadas YoY'] = vendas_anuais['Caixas Enviadas'].pct_change()
vendas_anuais['Valor Medio por Caixa YoY'] = vendas_anuais['Valor Medio por Caixa'].pct_change()
```

```python
display(vendas_anuais.style.format({
    'Valor': '${:,.2f}', # $1,000,000.00
    'Caixas Enviadas': '{:,.0f}',
    'Valor Medio por Caixa': '${:,.2f}',
    'Valor YoY': '{:,.1%}',
    'Caixas Enviadas YoY': '{:,.1%}', # 5,2%
    'Valor Medio por Caixa YoY': '{:,.1%}'
}))
```

##### Análise de produtos por ano

```python
df_produtos_anual = vendas_df[['Ano', 'Produto', 'Valor']]

# Agrupa por Ano e Produto
df_produtos_anual = df_produtos_anual.groupby(['Ano', 'Produto'], as_index=False).sum()
```

```python
# color='Produto' adiciona cores distintas para cada produto
grafico_produtos = px.bar(df_produtos_anual, x='Ano', y='Valor', color='Produto')
grafico_produtos.show()
```
